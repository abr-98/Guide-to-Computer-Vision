```python
import cv2
```

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
image=cv2.imread("cameraman1.png")
```

```python
plt.imshow(image)
```

```python
## Logarithmic stretching
```

```python
max_1=np.max(image)
```

```python
constant=255/(np.log(1 + max_1))
```

```python
image_l=constant*np.log(1+image)
```

```python
image_l = image_l.astype(int)
```

```python
plt.imshow(image_l)
```

```python
### Power law transform
```

```python
gamma=0.5
```

```python
image_p=constant*np.log(image+0.2)**gamma
```

```python
image_p = image_p.astype(int)
```

```python
plt.imshow(image_p)
```

```python
gamma=0.9
```

```python
image_p=constant*np.log(image)**gamma
```

```python
image_p = image_p.astype(int)
```

```python
plt.imshow(image_p)
```

```python
gamma=1.5
```

```python
image_p=constant*np.log(image)**gamma
```

```python
image_p = image_p.astype(int)
```

```python
plt.imshow(image_p)
```
