```python
### Min Max Stretching
```

```python
image=cv2.imread("hazy.jpg")
```

```python
plt.imshow(image)
```

```python
max=np.max(image)
min=np.min(image)
```

```python
image_cs=(image-min)/(max-min)*255
```

```python
image_cs = image_cs.astype(int)
```

```python
plt.imshow(image_cs)
```
