```python
import cv2
```

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
#### Prewitt Edge detector
```

```python
image=cv2.imread("cameraman1.png")
```

```python
plt.imshow(image)
```

```python
Prewitt_vertical_kernel=np.array(([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]), np.float32)
Prewitt_horizontal_kernel=np.array(([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]), np.float32)
```

```python
Prewitt_vertical_sharpen_img = cv2.filter2D(src=image, kernel=Prewitt_vertical_kernel, ddepth=-1)
```

```python
Prewitt_horizontal_sharpen_img = cv2.filter2D(src=image, kernel=Prewitt_horizontal_kernel, ddepth=-1)
```

```python
plt.imshow(Prewitt_vertical_sharpen_img)
```

```python
plt.imshow(Prewitt_horizontal_sharpen_img)
```

```python
### Laplacian of Gaussian Operator or Marr Hildreth Edge Detector
```

```python
## First Blur then apply Laplacian 
```

```python
blur_img = cv2.GaussianBlur(image,(3,3),0)
```

```python
laplacian = cv2.Laplacian(blur_img,cv2.CV_64F)
laplacian1 = laplacian/laplacian.max()
```

```python
plt.imshow(laplacian1)
```

```python
### Canny edge detector
```

```python
canny_d_img = cv2.Canny(image,20,100)
```

```python
plt.imshow(canny_d_img)
```
