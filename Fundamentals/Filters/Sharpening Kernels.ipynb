```python
import cv2
```

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
### Laplacian Sharpening filters
```

```python
## using 4 nieghbours
```

```python
image=cv2.imread("cameraman1.png")
```

```python
plt.imshow(image)
```

```python
Laplacian_4_kernel=np.array(([[0, -1, 0], [-1, 4, -1], [0, -1, 0]]), np.float32)
```

```python
Laplacian_4_sharpen_img = cv2.filter2D(src=image, kernel=Laplacian_4_kernel, ddepth=-1)
```

```python
plt.imshow(Laplacian_4_sharpen_img)
```

```python
### Sobel filters:
```

```python
Sobel_vertical_kernel=np.array(([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]), np.float32)
Sobel_horizontal_kernel=np.array(([[1, 2, 1], [0, 0, 0], [-1, -2, -1]]), np.float32)
```

```python
Sobel_vertical_sharpen_img = cv2.filter2D(src=image, kernel=Sobel_vertical_kernel, ddepth=-1)
```

```python
Sobel_horizontal_sharpen_img = cv2.filter2D(src=image, kernel=Sobel_horizontal_kernel, ddepth=-1)
```

```python
plt.imshow(Sobel_vertical_sharpen_img)
```

```python
plt.imshow(Sobel_horizontal_sharpen_img)
```
