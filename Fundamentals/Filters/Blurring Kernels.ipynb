```python
import cv2
```

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
### Mean Blur Kernel: Smoothening Kernel
```

```python
image=cv2.imread("cameraman1.png")
```

```python
plt.imshow(image)
```

```python
meanBlurKernel = np.ones((3, 3), np.float32)/9
```

```python
meanBlurimg = cv2.filter2D(src=image, kernel=meanBlurKernel, ddepth=-1)
```

```python
plt.imshow(meanBlurimg)
```

```python
### Weighted Averaging filter
```

```python
Weighted_mean_kernel=np.array(([[1, 2, 1], [2, 4, 2], [1, 2, 1]]), np.float32)/16
```

```python
Weightedmeanblurimg = cv2.filter2D(src=image, kernel=Weighted_mean_kernel, ddepth=-1)
```

```python
plt.imshow(Weightedmeanblurimg)
```

```python
### Guassian Blur Kernel
```

```python
from numpy import pi, exp, sqrt
s, m, k = 4,0, 1 #  generate a (2k+1)x(2k+1) gaussian kernel with mean=0 and sigma = s
probs = [exp(-(z-m)*(z-m)/(2*s*s))/s*sqrt(2*pi) for z in range(-k,k+1)] 
kernel = np.outer(probs, probs)
```

```python
#s=sigma, m=mean, k=dimension
```

```python
### Guassian distribution= (e^(-1/2*(x-mean)^2/sigma^21))*(1/(sigma*(2*pie)^(1/2))
```

```python
kernel
```

```python
Gaussianmeanblurimg = cv2.filter2D(src=image, kernel=kernel, ddepth=-1)
```

```python
plt.imshow(Gaussianmeanblurimg)
```
